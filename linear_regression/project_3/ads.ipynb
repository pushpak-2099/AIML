```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Step_1 -- importing data set
df=pd.read_csv('Advertising.csv')
print(df.head())

#trying to work on simple linear regression 
#dropping other columns
print(df.columns)
df=df.drop(['Unnamed: 0',  'Radio', 'Newspaper'],axis=1)

print(df.head())
print(df.info())
print(df.describe())
```

```python
#Step_2 -- selecting the features
X=df[["TV"]] #2d array
y=df["Sales"]


#Step_3 -- train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=50)
```

```python
#Step_4 -- Standardrization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)

```

```python
#Step_5 -- regression model
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test)
```

```python
#visualization
plt.scatter(X_train,y_train)
plt.plot(X_train_scaled,linreg.predict(X_train_scaled),"r")

print("coefficient :",linreg.coef_)
print("intercept :",linreg.intercept_)
```

```python
#check the preformance of model
y_pred_test=linreg.predict(X_test)
print(y_pred_test,"\n", y_test)
```

```python
#for finding the accuracy
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred_test)
print("mse =",mse)
mae=mean_absolute_error(y_test,y_pred_test)
print("mae =",mae)
rmse=root_mean_squared_error(y_test,y_pred_test)
print("rmse =",rmse)
score=r2_score(y_test,y_pred_test)
print("r2 score =",score)

y_pred=linreg.predict(X_train_scaled)
r2_train=r2_score(y_train,y_pred)
print("r2 score for train data =",r2_train)
```
